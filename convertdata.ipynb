```python
import pandas as pd
import json

# Load CSV
df = pd.read_excel(r"C:\Users\isabellamyssior\OneDrive - Statista GmbH\Documents\Revenue Engineering\Statista overview\Data Points.xlsx")

# Clean column names (just in case)
df.columns = df.columns.str.strip()

# Create dictionary
result = {}

for _, row in df.iterrows():
    topic = row["Topic"]
    name = row["Chapter Name"]
    value = row["Data Points"]

    # Convert value to int (removes commas if needed)
    try:
        value = int(str(value).replace(",", ""))
    except:
        value = 0

    if topic not in result:
        result[topic] = []

    result[topic].append({
        "name": name,
        "value": value
    })

# Save JSON
with open("data.json", "w") as f:
    json.dump(result, f, indent=2)

print("✅ data.json created successfully")
```
